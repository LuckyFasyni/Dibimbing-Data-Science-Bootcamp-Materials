# Using Subplots

$\textbf{Now how if we want to make a single plot in a lot of axes?}$

This is where $\textbf{subplots}$ does its job. The documentation can be seen [here](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplots.html)

Create a figure and a set of subplots. This utility wrapper makes it convenient to create common layouts of subplots, including the enclosing figure object, in a single call.

# One Axes

# Multiple Axes

Pass the following variables as keyword args: x, y. From version 0.12, the only valid positional argument will be `data`, and passing other arguments without an explicit keyword will result in an error or misinterpretation. warnings.warn(. hiraukan aja ya? 

```python
fig,axes = plt.subplots(2, # number of rows
                        1, # number of columns
                        figsize=(8,5), # ukuran figuranya
                        sharex=True, sharey=True
                        )

axes[0].plot(year,business)
axes[1].plot(year,physical_sciences, color='r')
```

```python
fig,axes = plt.subplots(3, # number of rows
                        2, # number of columns
                        figsize=(8,5), # ukuran figuranya
                        )

axes[1,0].plot(year,business)
```

```python
fig,axes = plt.subplots(2, # number of rows
                        1, # number of columns
                        figsize=(8,5), # ukuran figuranya
                        sharex=True, sharey=True
                        )

axes[0].plot(year,physical_sciences,color='blue')
axes[0].plot(year,computer_sciences,color='orange')

axes[1].plot(year, business, color='red')
axes[1].plot(year, psychology, color='purple')
```

# Add Title

```python
plt.plot(year, physical_sciences, color='blue')
```

```python
plt.figure(figsize= (8,5))
plt.plot(year, physical_sciences, color='blue')

plt.title('Physical Sciences Bachelors over Years', fontsize=16)
plt.show()
```

```python
fig, axes = plt.subplots(2, 1, figsize=(8,5))

axes[0].plot(year,physical_sciences, color='blue')
axes[1].plot(year,computer_sciences, color='red')

axes[0].set_title('Physical Science Bachelors over Years', fontsize=16)
axes[1].set_title('Computer Science Bachelors over Years', fontsize=16)

fig.tight_layout(pad=1)
```

# Add Label

```python
plt.figure(figsize= (8,5))
plt.plot(year, physical_sciences, color='blue')
plt.title('Physical Sciences Bachelors over Years', fontsize=16)

# add label
plt.xlabel('Year', fontsize=12)
plt.ylabel('Percentage', fontsize=12)

plt.show()
```

```python
fig, axes = plt.subplots(2, 1, figsize=(8,5), sharex=True, sharey=True)

axes[0].plot(year,physical_sciences, color='blue')
axes[1].plot(year,computer_sciences, color='red')

## add x label di axes ke-2
axes[1].set_xlabel('Year', fontsize=12)

## add y label
# axes[0].set_ylabel('Percentage', fontsize=12)
# axes[1].set_ylabel('Percentage', fontsize=12)
fig.text(0.05,0.5,'Percentage', rotation = 'vertical', fontsize=12)
```
