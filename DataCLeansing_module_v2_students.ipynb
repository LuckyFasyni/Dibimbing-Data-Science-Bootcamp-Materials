## 7. Other Data Preprocessing

Label Encoding

```python
from sklearn.preprocessing import OrdinalEncoder

ord_enc = OrdinalEncoder()
```

```python
df_encode = pd.DataFrame(df['country'])
```

```python
df_encode['countryCode'] = ord_enc.fit_transform(df_encode[['country']])
```

```python
df_encode
```

> One Hot Encoding

```python
df_country = df[['country']]
```

```python
df_country
```

```python
df_country_encoded = pd.get_dummies(df_country,columns=['country'],prefix='country')
```

```python
df_country_encoded.head()
```

```python
pd.concat([df_country,df_country_encoded],axis=1)
```

> Normalization

```python
df.head()
```

```python
df['ordervalue'] = pd.to_numeric(df['ordervalue'].str.replace(',',''))
```

```python
from sklearn.preprocessing import MinMaxScaler

min_max_scaler = MinMaxScaler()
```

```python
df['ordervalue_norm'] = min_max_scaler.fit_transform(df[['ordervalue']])
```

```python
df.describe()
```

> Standardization

```python
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()
```

```python
df['ordervalue_std'] = std_scaler.fit_transform(df[['ordervalue']])
```

```python
df.head()
```
