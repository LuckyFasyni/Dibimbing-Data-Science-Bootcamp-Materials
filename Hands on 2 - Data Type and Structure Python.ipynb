### Basic Operation with Boolean

There are several basic calculation you can do with the boolean type of data, including `or`, `and`, `not`.

### String

```python
x = 'Louis'
y = "Owen"
```

If you want to create a multi-line string, you can use `'''` or `"""`.

```python
# Append 2 strings
print(x+y)
```

```python
# Repeat string
print(x * 3)
```

```python
# Index Slicing
print(x[2] + x[-1])
```

```python
# String Lowering
print(x.lower())
```

```python
# String Capitalizing
print(x.upper())
```

```python
# Check whether the string contains only digit
print(x.isdigit())
print('1'.isdigit())
print('bt21'.isdigit())
```

```python
# Substring replacement
print(x.replace('uis','wen'))
```

```python
# String Splitting
text = 'Python is a fun programming language. It is also a very powerful language for Data Science.'

print(text.split())
```

```python
print(text.split(' '))
```

```python
print(text.split('.'))
```

## Non-Primitive Data Structures

### List

List is a data structure in python that can saves many different types of other items. 

```python
lst_1 = ['a','b','c','d','e']
print(lst_1)
```

```python
lst_2 = ['a','a', 1, True, 4.3]
print(lst_2)
```

### Basic Operations with List

```python
# Appending 2 Lists
print(lst_1 + lst_2)
```

```python
# Range Slicing
print(lst_1[1:3])
print(lst_1[-4:])
print(lst_1[4:])
```

Note that although we sliced from index 1 to 3, but the sliced list is only from index 1 to 2. Why? Because the ending parameter of slicing in list is exclusive (have to be the index after the expected ending index).

```python
# Accessing Specific Value
print(lst_1[3])
```

```python
# Replacing Specific Value
lst_1[3] = 'z'
print(lst_1)
```

```python
# Replacing Ranges of Values
lst_1[3:] = ['d','e']
print(lst_1)
```

```python
# Appending New Value at the end of the list
lst_1.append('f')
print(lst_1)
```

```python
# Removing Last Value of the list
lst_1.pop()
print(lst_1)
```

```python
# Removing Value at Specific Index of the list
lst_1.pop(2)
print(lst_1)
```

```python
# Inserting New Value at Specific Index
lst_1.insert(2,'z')
print(lst_1)
```

```python
# Removing the first occurence of specific value from the list
lst_1.remove('z')
print(lst_1)
```

```python
# Reverse the List Order
print(lst_1[::-1])
```

```python
# Checking if a value is contained in the list
'a' in lst_1
```

```python
# Filtering Values that contained in lst_1 but not in lst_2
[x for x in lst_1 if x not in lst_2]
```

```python
# Join list elements into string
print(''.join(lst_1))
print(','.join(lst_1))
```

This operations actually shows us that **a string is actually a list of character**.

### Arrays

Array is very similar to list. The only different is array can only contains the same type of data. However, this limitation turns out enables us to perform an operations on **all its item individually** easily, which may not be the case with lists.

The most popular implementation of array in Python can be found in the **Numpy** package.

```python
import numpy as np
```

```python
arr_1 = np.array([1,2,3,4])
print(arr_1)
```

What if we force it to contains different types of elements?

```python
arr_mixed_1 = np.array(['a','b',1,2])
print(arr_mixed_1)
```

It turns out, all of the elements in the array is converted into string. Why? Becuase the type will be determined as the **minimum type** required to hold the objects in the array.

```python
arr_mixed_2 = np.array([1,2,True])
print(arr_mixed_2)
```

```python
arr_mixed_2 = np.array([1.5,2,True])
print(arr_mixed_2)
```

There's also a function from Numpy to generate a specific type of array.

```python
arr_ones = np.ones(10)
print(arr_ones)
```

```python
arr_zeros = np.zeros(10)
print(arr_zeros)
```

### Basic Operations with Numpy Array

Unlike Lists, array performs the operation in the element-wise fashion. It means that the operation is performed in each of the corresponding elements index.

```python
arr_1 = np.ones(10)
print(arr_1)

arr_2 = 5 * np.ones(10)
print(arr_2)
```

```python
# Addition
print(arr_1 + arr_2)
```

```python
# Substraction
print(arr_1 - arr_2)
```

```python
# Multplication
print(arr_1 * arr_2)
```

```python
# Division
print(arr_1 / arr_2)
```

```python
# Modulo
print(arr_1 % arr_2)
```

```python
# Floor Divison
print(arr_1 // arr_2)
```

```python
# Power / Exponentiation
print(arr_2 ** arr_2)
print(arr_2 ** 0.5) # square root
```

```python
arr_3 = np.array([1,20,3,5,0.4,2,9,6,20])
```

```python
# Average over array
avg_arr_3 = np.mean(arr_3)
print(avg_arr_3)
```

```python
# Median over array
median_arr_3 = np.median(arr_3)
print(median_arr_3)
```

```python
# Minimum over array
minimum_arr_3 = np.min(arr_3)
print(minimum_arr_3)
```

```python
# Maximum over array
maximum_arr_3 = np.max(arr_3)
print(maximum_arr_3)
```

### Tuple

Tuple is another data structures that is similar with list. However, a tuple is not immutable, which means once it is defined, you can not delete, add or edit any values inside it. 

This characteristic of tuple is useful when you want to make sure that the information you have stored and will be transferred to other people can only be viewed and accessed, but can not be edited.

```python
tuple_1 = (1,2,3)
print(tuple_1)
```

```python
tuple_2 = 4,5,6
print(tuple_2)
```

```python
print(tuple_1[0])
```

```python
print(tuple_1 + tuple_2)
```

### Set

Set is another list-alike data structure. However, set contains only unique values and it is not ordered.

```python
set_1 = {1,2,3,4,4,4,4,4,4}
print(set_1)
```

```python
set_2 = set([4,4,4,4,4,5,6,7])
print(set_2)
```

### Basic Operations with Set

```python
# All the elements in set_1 but not in set_2
print(set_1 - set_2)
```

```python
# Elements in set_1 or set_2 or both (UNION)
print(set_1 | set_2)
```

```python
# Elements in both set_1 and set_2 (INTERSECTION)
print(set_1 & set_2)
```

### Dictionary

Dictionary is a data structure that enables us to have a direct access for specific value via the corresponding key. You can imagine this data structure just like the actual dictionary.

A dictionary contains of the `key-value` pairs. Key is responsible as the identifier on how to access the value. Here, the value can be any types of data, whether it is a list, string, integer, set, or even a dictionary.

```python
# Let say we want to create a dictionary of test scores for each of the students
score_dict = {'Joe':70,
            'Thom':95,
            'Monica':80,
            'Max':65
            }
print(score_dict)
```

```python
# We can access the score for each of the students via its key (name of the student)
print(score_dict['Joe'])
print(score_dict['Max'])
```

```python
# We can add the element if there's new entry 
score_dict['Stella'] = 76
print(score_dict)
```

```python
# We can change the value of previous entry
score_dict['Stella'] = 80
print(score_dict)
```

```python
# We can also delete a specific entry
del score_dict['Stella']
print(score_dict)
```

```python
# We'are also able to access the sequences of our dictionary keys and values
print(score_dict.keys())
print(score_dict.values())
```
